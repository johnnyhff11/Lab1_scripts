### Fourier Transform of the alia_real data

```python
fft_alia800_1100 = np.fft.fft(alia800_1100['arr_0'])
freq_fft = np.fft.fftfreq(2048)
x1 = np.linspace(-1024, 1024, 2048)
fft_freq = np.fft.fftfreq(2048, dt)
realfft = fft_alia800_1100.real
imagfft = fft_alia800_1100.imag
x2= np.fft.fftfreq(2048, dt)
x = np.fft.fftshift(x2)

plt.figure()
plt.plot(x,np.fft.fftshift(fft_alia800_1100), alpha = 0.75, label = 'real&imaginary')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Voltage (mV)')
plt.title('Fourier Spectrum of 800kHz Wave Sampled at 1.1mHz')
plt.legend()
plt.show()

plt.figure()
plt.plot(x,np.fft.fftshift(realfft), alpha = 0.5, label = 'real')
plt.plot(x,np.fft.fftshift(imagfft), alpha = 0.5, label = 'imag')
plt.legend()
plt.show()
```

### Power Spectrum of alia_real

```python
pwr_alia800_1100 = abs(fft_alia800_1100)**2
pwr_realfft = abs(realfft)**2
pwr_imagfft = abs(imagfft)**2

plt.figure()
plt.plot(x, np.fft.fftshift(pwr_alia800_1100), label='Real&Imag')
plt.xlabel('Frequency (Hz)')
plt.legend()
plt.ylabel('Relative Power')
plt.title('Power Spectrum of 800kHz Wave Sampled at 1.1mHz')
plt.show()

plt.figure()
plt.plot(x,np.fft.fftshift(pwr_realfft), alpha=0.5, label='Real')
plt.plot(x,np.fft.fftshift(pwr_imagfft), alpha=0.5, label='Imag')
plt.legend()
plt.xlabel('Frequency (Hz)')
plt.ylabel('Relative Power')
plt.show()
```

### Nyquist Windows

```python
samp_freq = 1.1e6


plt.figure()
plt.plot(x, np.fft.fftshift(fft_alia800_1100), color='lime', label='Fourier Spectrum')
# plt.axvline(x=samp_freq, linestyle='dashed', color='red', label='Nyquist Bounds')
plt.axvline(x=samp_freq/2, linestyle='dashed', color='red', label='Nyquist Bound $ {v_{samp}} / {2} $')
plt.xlim()
plt.axvline(x=(samp_freq)/4, linestyle='dashed', color='red')
plt.legend()
plt.title('800kHz Sine Wave Sampled at 1.1mHz with half of Nyquist rate')
plt.show()
```

### Comparing the aliasing and non-aliasing components

```python
non_alia800_2200 = np.load('alia_800kHz_22db.npz')
```

```python
samprate22 = 2.2e6
time = np.linspace(0,2048/2.2e6, 2048)
dt = time[1] - time[0]
prox = np.fft.fftfreq(2048, dt)
x22 = np.fft.fftshift(prox)


fft_nonalia800_2200 = np.fft.fft(non_alia800_2200['arr_0'])
shift_nonalia800_2200 = np.fft.fftshift(fft_nonalia800_2200)

plt.figure()
plt.plot(x22,shift_nonalia800_2200, label='Fourier Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Voltage (mV)')
plt.title('Fourier Spectrum of 800kHz Sampled at 2.2mHz')
plt.legend()
plt.show()

plt.figure()
pwr_nonalia800_2200 = abs(shift_nonalia800_2200)**2
plt.plot(x22, pwr_nonalia800_2200, label='Power Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Relative Power')
plt.title('Power Spectrum of 800kHz Sampled at 2.2mHz')
plt.legend()
plt.show()
```
