```python
import numpy as np
import matplotlib.pyplot as plt
import ugradio
```

```python
data = np.load("aliased_data(400kHz,100mV,3attenuator).npz")
```

```python
data.files
```

```python
x= data["arr_0"]
```

```python
y = data["arr_1"]
```

```python
plt.figure()
plt.plot(y,x)
plt.xlim(0, 10)
plt.show()
```

```python
fft = np.fft.fft(data["arr_0"])
```

```python
plt.figure()
plt.plot(abs(fft))
plt.show() #This is FT
```

```python
fft_inv = np.fft.fftshift(fft)
#fft_inv = fft[1024]
#fft.shape
x1 = np.linspace(-1024, 1024, 2048)
```

```python
plt.figure()
plt.plot(x1, abs(fft_inv))
plt.show()
```

```python
power = abs(fft_inv)**2
```

```python
plt.figure()
plt.plot(x1, power)
plt.show()
```

```python
ima = fft_inv.imag
rea = fft_inv.real
```

```python

plt.plot(ima)
plt.plot(rea)
plt.savefig('Forrest\'s Voltage Spectra.pdf', dpi=150)
plt.show()
```
