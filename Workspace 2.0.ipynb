```python
import numpy as np
import matplotlib.pyplot as plt
import ugradio as ug
from numpy import load
import scipy as sp
from scipy import stats
from scipy.stats import norm
import scipy as sp
import scipy.signal as sps
```

```python
fft_ls = []
pwr_ls =[]
def fourier(array):
    for i in range(10):
        a = array[i]
        b = np.fft.fft(a)
        fft_ls.append(b)
    return fft_ls

def power(fft):
    for i in range(len(fft)):
        a = fft[i]
        b = np.abs(a)**2
        pwr_ls.append(b)
    return pwr_ls
```

## Starting with 400kHz

```python
#Aliased
aliased400 = np.load("aliased_data(400kHz,100mV,3attenuator).npz")
alia_400 = aliased400["arr_0"]
x = aliased400["arr_1"]
alia_400_fft = np.fft.fft(alia_400)
pwr_alia_400 = np.abs(alia_400_fft)**2

#Antialiasing 
anti400 = np.load("anti_data(400kHz,100mV,3attenuator).npz")
anti_400 = anti400["arr_0"]

anti_400_fft = np.fft.fft(anti_400)
pwr_anti_400 = np.abs(anti_400_fft)**2
##### PLOTTING #####

fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(10,8))
ax1.plot(x,alia_400_fft)
ax1.set_title("Voltage spectra 400kHz aliased")
ax1.set_xlabel("frequency kHz")
ax1.set_ylabel("voltage mV")

ax2.plot(x,pwr_alia_400,color='red')
ax2.set_title("Power Sprectrum 400 kHz aliased")
ax2.set_xlabel("frequency kHz")
ax2.set_ylabel("Power")

ax3.plot(x,anti_400_fft)
ax3.set_title("Voltage spectra 400kHz anti")
ax3.set_xlabel("frequency kHz")
ax3.set_ylabel("voltage mV")

ax4.plot(x,pwr_anti_400,color='red')
ax4.set_title("Power Sprectrum 400 kHz anti")
ax4.set_xlabel("frequency kHz")
ax4.set_ylabel("Power")

plt.show()
fig.savefig("400kHz.png")
```

```python
noise_10blocks = np.load("noise_data_10blocks.npz")
noise_data_10blocks = noise_10blocks["arr_0"]
noise_data_10blocks.shape


fn = fourier(noise_data_10blocks)
pn = power(fn)
plt.figure(figsize=(8,6))

for i in range(10):
    plt.plot(pn[i],label=f"nblock={i}")
    plt.legend()
    plt.show()
```
