# Retaking aliased vs anti waves for transforms and power 

```python
## 1.1MHz ##
aliased_1 = np.load("alia_1.1MHz_22db.npz")
alia_1 = aliased_1["arr_0"]
x0 = aliased_1["arr_1"]
x1 = np.fft.fftfreq(len(aliased_1["arr_1"]))
x2 = np.fft.fftshift(x1)
# x = aliased_1["arr_1"]
# x2 = np.fft.fftshift(x)
alia_1_fft = np.fft.fft(alia_1)
alia_1_ffts = np.fft.fftshift(alia_1_fft)
alia_1_pwr = np.abs(alia_1_fft)**2

fig,((ax1,ax2,ax3)) = plt.subplots(3,1,figsize=(10,10))
ax1.plot(alia_1[:50],color='red')
ax1.set_title("sine wave 1.1Mhz aliased")
ax1.set_xlabel("frequency kHz")
ax1.set_ylabel("voltage mV")

ax2.plot(x0,alia_1_fft,color='purple')
ax2.set_title("fourier transform 1.1MHz aliased")
ax2.set_xlabel("frequency kHz")
ax2.set_ylabel("Voltage mV")

ax3.plot(alia_1_pwr,color='gold')
ax3.set_title("Power Spectra 1.1Mhz aliased")
ax3.set_xlabel("frequency kHz")
ax3.set_ylabel("Power")

plt.show()
# print(alia_1_pwr[1023])
# print(alia_1_pwr[1024]) Peak here
# print(alia_1_pwr[1025])
# print(alia_1_pwr[1026])
# len(alia_1_pwr)
```

```python
## 1.1MHz ##
anti_1 = np.load("anti_1.1MHz_22db.npz")
anti_1 = aliased_1["arr_0"]
anti_1_fft = np.fft.fft(anti_1)
anti_1_ffts = np.fft.fftshift(anti_1_fft)
anti_1_pwr = np.abs(anti_1_fft)**2

fig,((ax1,ax2,ax3)) = plt.subplots(3,1,figsize=(10,10))
ax1.plot(anti_1[:100],color='red')
ax1.set_title("sine wave 1.1Mhz anti")
ax1.set_xlabel("frequency kHz")
ax1.set_ylabel("voltage mV")

ax2.plot(x2,anti_1_ffts,color='purple')
ax2.set_title("fourier transform 1.1MHz anti")
ax2.set_xlabel("frequency kHz")
ax2.set_ylabel("Voltage mV")

ax3.plot(x2,anti_1_pwr,color='gold')
ax3.set_title("Power Spectra 1.1Mhz anti")
ax3.set_xlabel("frequency kHz")
ax3.set_ylabel("Power")

plt.show()
```
