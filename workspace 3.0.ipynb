```python
import numpy as np
import matplotlib.pyplot as plt
import ugradio
from numpy import load
import scipy as sp
from scipy import stats
from scipy.stats import norm
import scipy as sp
import scipy.signal as sps
```

```python
mixer1 = np.load("mixer2.npz")
mixer1_data = mixer1["arr_0"]

plt.figure(figsize=(8,6))
plt.plot(mixer1_data,label='mixer1 750&650')
plt.xlim(0,250)
plt.legend()
plt.show()

mixer1_ffts = np.fft.fft(mixer1_data)
mixer1_fft = np.fft.fftshift(mixer1_ffts)
x1 = np.fft.fftfreq(len(mixer1["arr_1"]))
x2 = np.fft.fftshift(x1)

plt.figure(figsize=(8,6))
plt.plot(x2,mixer1_fft,label='mixer fft')
plt.legend()
plt.show()

plt.figure(figsize=(8,6))
plt.plot(mixer1_ffts,label='mixer fft non shifted')
plt.xlim(0,1024)
plt.grid()
plt.legend()
plt.show()
```
